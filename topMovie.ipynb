```python
from selenium import webdriver
```

```python
movieDriver = webdriver.Chrome()
```

```python
movieDriver.get('https://www.boxofficemojo.com/chart/top_lifetime_gross/?area=XWW')
```

```python
movies = movieDriver.find_elements("xpath","//tr[@style='display: none;']/following-sibling::tr")
```

```python
for movie in movies:
    print(movie.text)
```

```python
!pip install pandas
```

```python
import pandas as pd
from selenium.webdriver.common.by import By
```

```python
import csv

movie_list = []
for movie in movies:
    cells = movie.find_elements(By.TAG_NAME, 'td')
    if cells:
        row = [cell.text for cell in cells]
        movie_list.append(row)
        
```

```python
df = pd.DataFrame(movie_list, columns=['Rank', 'Title', 'Lifetime Gross', 'Year'])
df.to_csv("topMovies.csv", index=False) 
```

```python
movieDriver.quit()
```
